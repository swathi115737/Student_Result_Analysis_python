```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Student_scores.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum() ## to know the null values
```

## drop unnamed column

```python
df=df.drop("Unnamed: 0",axis=1)
df.head()
```

## data transformation or transformation of columns

## Change wkly study hours

```python
df["WklyStudyHours"] = df["WklyStudyHours"].str.replace("05-Oct","5-10")
df.head()
```

## Gender distrubution

```python
plt.figure(figsize= (4,3)) ## figure size ## breadth+height
sns.countplot(data=df, x="Gender")
plt.title("Gender distrubution") ## to give a title for chart
plt.show()
```

```python
plt.figure(figsize= (5,5)) ## figure size ## breadth+height
ax=sns.countplot(data=df, x="Gender") ## to know excat count 
ax.bar_label(ax.containers[0])
plt.show()
```

```python
## from the above chart we have analysed that:
## the number of females in the data is more than the number of males

gb=df.groupby("ParentEduc").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
print(gb)
```

```python
plt.figure(figsize= (5,5)) ## to decrease the size
sns.heatmap(gb, annot = True) ## we can represent analysis brefily, annot=True is used to get values in table
plt.title("Relationship between parent's Education and student's score")
plt.show()
```

```python
# From the above chart we have concluded that the education of the parents have a good impact on their scores

gb1=df.groupby("ParentMaritalStatus").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"}) ## to know whether ParentMaritalStatus is impacting on students or not
print(gb1)
```

```python
plt.figure(figsize= (5,5)) ## to decrease the size
sns.heatmap(gb1, annot = True) ## we can represent analysis brefily, annot=True is used to get values in table
plt.title("Relationship between Parent's Marital Status and Student's score")
plt.show()
```

```python
## from the above chart we have concluded that there is no/negligible impact on the student's score on their ParentMaritalStatus
```

## Distribution of Ethnic Groups

```python
groupA = df.loc[(df['EthnicGroup'] == "group A")].count() ## to know groupA values
print(groupA)
```

```python
groupA = df.loc[(df['EthnicGroup'] == "group A")].count()
groupB = df.loc[(df['EthnicGroup'] == "group B")].count()
groupC = df.loc[(df['EthnicGroup'] == "group C")].count()
groupD = df.loc[(df['EthnicGroup'] == "group D")].count()
groupE = df.loc[(df['EthnicGroup'] == "group E")].count()

l=["group A", "group B", "group C", "group D", "group E"] ## labels
mlist = [groupA["EthnicGroup"], groupB["EthnicGroup"], groupC["EthnicGroup"], groupD["EthnicGroup"], groupE["EthnicGroup"]]
plt.pie(mlist, labels=l, autopct="%1.2f%%") ## autopct is used to see the percentageplt.
plt.title("Distribution of Ethnic Groups")
plt.show()
```
